### Hints: [隨機產生數值](https://blog.csdn.net/christianashannon/article/details/78867204)

```python
import pandas as pd
import numpy as np
import random
import string
```

```python
def rand_country(N=10):
    def _gen():
        return ''.join(random.choice(string.ascii_uppercase + string.digits) for _ in range(5))
    return [_gen() for i in range(N)]

data = {'國家': rand_country(50),
        '人口': np.random.randint(low=1000000, high=100000000, size=50)}
data = pd.DataFrame(data)
data
```

```python
data.loc[data['人口'].idxmax()]
```
